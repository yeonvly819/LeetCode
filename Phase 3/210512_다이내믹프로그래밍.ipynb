```python
x = int(input())

d = [0] * 30

for i in range(2, x + 1):
    d[i] = d[i-1] + 1
    print('prev d:', d)
    print('d[', i, ']is:', d[i])
    print('d[', i-1, ']is:', d[i-1])

    if i % 2 == 0:
        print('d[i]:', d[i])
        print('d[i//2]', d[i//2])
        d[i] = min(d[i], d[i//2] + 1)
    
    if i % 3 == 0:
        print('d[i]:', d[i])
        print('d[i//2]', d[i//3])
        d[i] = min(d[i], d[i//3] + 1)
    
    if i % 5 == 0:
        print('d[i]:', d[i])
        print('d[i//2]', d[i//5])
        d[i] = min(d[i], d[i//5] + 1)
    
    print('post d:', d)
    print('=====')

print(d[x])
```
